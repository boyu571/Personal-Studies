```python
CA = raw_df.groupby('Subject').get_group('Current account, balance')
FA = raw_df.groupby('Subject').get_group('Financial account, net')
FDI_assets = raw_df.groupby('Subject').get_group('Direct investment, net acquisition of financial assets')
FDI_liabilities = raw_df.groupby('Subject').get_group('Direct investment, net incurrence of liabilities')
Portfolio_assets = raw_df.groupby('Subject').get_group('Portfolio investment, net acquisition of financial assets')
Portfolio_liabilities = raw_df.groupby('Subject').get_group('Portfolio investment, net incurrence of liabilities')
derivatives = raw_df.groupby('Subject').get_group('Financial derivatives, net')
Portfolio = raw_df.groupby('Subject').get_group('Portfolio investment, net')
Other = raw_df.groupby('Subject').get_group('Other investment, net')
FDI = raw_df.groupby('Subject').get_group('Direct investment, net')
Reserve = raw_df.groupby('Subject').get_group('Reserve assets, net acquisition of financial assets')
```

```python
CA = CA.drop(columns="Subject")
FA = FA.drop(columns="Subject")
FDI = FDI.drop(columns="Subject")
Portfolio = Portfolio.drop(columns="Subject")
derivatives = derivatives.drop(columns="Subject")
Other = Other.drop(columns="Subject")
Reserve = Reserve.drop(columns="Subject")
FDI_assets = FDI_assets.drop(columns="Subject")
FDI_liabilities = FDI_liabilities.drop(columns="Subject")
Portfolio_assets = Portfolio_assets.drop(columns="Subject")
Portfolio_liabilities = Portfolio_liabilities.drop(columns="Subject")
```

```python
CA.columns = ["TIME", "CA"]
FA.columns = ["TIME", "FA"]
FDI.columns = ["TIME", "FDI"]
Portfolio.columns = ["TIME", "Portfolio"]
derivatives.columns = ["TIME", "derivatives"]
Other.columns = ["TIME", "Other"]
Reserve.columns = ["TIME", "Reserve"]
FDI_assets.columns = ["TIME", "FDI_assets"]
FDI_liabilities.columns = ["TIME", "FDI_liabilities"]
Portfolio_assets.columns = ["TIME", "Portfolio_assets"]
Portfolio_liabilities.columns = ["TIME", "Portfolio_liabilities"]
```

```python
from functools import reduce
```

```python
data_list = [CA, FA, FDI, Portfolio, derivatives, Other, Reserve, FDI_assets, FDI_liabilities, Portfolio_assets, Portfolio_liabilities]
df_total = reduce(lambda x, y : pd.merge(x,y,on='TIME'), data_list)
```

```python
df_total.to_csv('dataset1.csv')
```
